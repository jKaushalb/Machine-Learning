```
import sklearn
from sklearn import metrics,datasets
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn import model_selection
```

```
# spliting data
X,Y=datasets.load_iris(return_X_y=True)

xtrain=X[range(0,150,2),:]
ytrain=Y[range(0,150,2)]
xtest=X[range(1,150,2),:]
ytest=Y[range(1,150,2)]

# making instance of id3 classifier
print("Classification Using ID3 ")
clf=DecisionTreeClassifier(criterion='entropy')
#learning the classifier
clf.fit(xtrain,ytrain)
#predicting the class
ypred=clf.predict(xtest)

print("accuracy ",metrics.accuracy_score(ytest,ypred))
print("confussion metrics ",metrics.confusion_matrix(ytest,ypred))
print()
print(metrics.classification_report(ytest, ypred))
print()

# CART 

print("classification using CART ")
# making instance of cart classifier
model = DecisionTreeClassifier(criterion='gini')
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
print("accuracy ",metrics.accuracy_score(ytest,ypred))
print("confussion metrics ",metrics.confusion_matrix(ytest,ypred))
print()
print(metrics.classification_report(ytest, ypred))
print()
```

```
import pandas as pd
import csv
import random 
from sklearn import preprocessing
from sklearn.metrics import classification_report 
```

```
trend = 'up,down,down,down,up,up,down,up,up,up,down,down,down,up,down,up,down,up,up,down'.split(',')
variance = 'low,high,low,normal,high,low,normal,high,normal,low,normal,normal,high,high,low,high,high,normal,normal,low'.split(',')
buys = 'yes,no,no,no,yes,yes,yes,no,yes,yes,no,yes,no,yes,yes,no,no,yes,yes,no'.split(',')
table={'trend':trend,'variance':variance,'Purchase':buys}
df = pd.DataFrame(table)
df


```

```
label_encoder = preprocessing.LabelEncoder()

for i in df:
  df[i] = label_encoder.fit_transform(df[i])
print("After preprocessing ")
df
```

```
data = df[['trend','variance']]
out = df['Purchase']
x_train, x_test, y_train, y_test = model_selection.train_test_split(data, out, test_size = 0.30)
print(x_train)
```

```
print("For ID3\n")
clf = DecisionTreeClassifier(criterion = 'entropy')

clf.fit(x_train, y_train)
ypred = clf.predict(x_test)
print(ypred)
print('Accuracy', metrics.accuracy_score(y_test, ypred, normalize = True))

print("For CART\n")
clf = DecisionTreeClassifier(criterion = 'gini')
clf.fit(x_train, y_train)
ypred = clf.predict(x_test)

print(ypred)
print('Accuracy', metrics.accuracy_score(y_test, ypred, normalize = True))
```
